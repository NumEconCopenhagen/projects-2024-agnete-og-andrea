# ***Question 1***

```python
par = model.par
# a. total endowment
w1bar = 1.0
w2bar = 1.0

# b. figure set up
fig = plt.figure(frameon=False,figsize=(6,6), dpi=100)
ax_A = fig.add_subplot(1, 1, 1)

ax_A.set_xlabel("$x_1^A$")
ax_A.set_ylabel("$x_2^A$")

temp = ax_A.twinx()
temp.set_ylabel("$x_2^B$")
ax_B = temp.twiny()
ax_B.set_xlabel("$x_1^B$")
ax_B.invert_xaxis()
ax_B.invert_yaxis()

# A
ax_A.scatter(par.w1A,par.w2A,marker='s',color='black',label='endowment') #plots the endowment 
xes = [] #creates empty list
yes = [] #creates empty list
for x in model.Cset():
    xes.append(x[0]) #In each iteration of the loop, these lines extract the first and second elements of the current data point x (assuming x is a tuple or list) and append them to the xes and yes lists, respectively. This effectively separates the x and y coordinates of each data point and stores them in separate lists.
    yes.append(x[1])
ax_A.scatter(xes,yes,marker="s", color='green',label='Pareto Pairs') #plots the xes list on the x-axis and yes list on the y-axis.

# limits
ax_A.plot([0,w1bar],[0,0],lw=2,color='black')
ax_A.plot([0,w1bar],[w2bar,w2bar],lw=2,color='black')
ax_A.plot([0,0],[0,w2bar],lw=2,color='black')
ax_A.plot([w1bar,w1bar],[0,w2bar],lw=2,color='black')

#adjusts the limits to make them pretty
ax_A.set_xlim([-0.1, w1bar + 0.1])
ax_A.set_ylim([-0.1, w2bar + 0.1])    
ax_B.set_xlim([w1bar + 0.1, -0.1])
ax_B.set_ylim([w2bar + 0.1, -0.1])

#Adds the describing legend to the figure:
ax_A.legend(frameon=True,loc='upper right',bbox_to_anchor=(1.6,1.0));
```

***b. Illustrate and compare with question 3-5***

We plot the allocations to compare them:

```python
par = model.par
# a. total endowment
w1bar = 1.0
w2bar = 1.0

# b. figure set up
fig = plt.figure(frameon=False,figsize=(6,6), dpi=100)
ax_A = fig.add_subplot(1, 1, 1)

ax_A.set_xlabel("$x_1^A$")
ax_A.set_ylabel("$x_2^A$")

temp = ax_A.twinx()
temp.set_ylabel("$x_2^B$")
ax_B = temp.twiny()
ax_B.set_xlabel("$x_1^B$")
ax_B.invert_xaxis()
ax_B.invert_yaxis()

# A
ax_A.scatter(par.w1A,par.w2A,marker='s',color='black',label='endowment') #plots the endowment 

xes = [] #creates empty list
yes = [] #creates empty list
for x in model.Cset():
    xes.append(x[0]) #In each iteration of the loop, these lines extract the first and second elements of the current data point x (assuming x is a tuple or list) and append them to the xes and yes lists, respectively. This effectively separates the x and y coordinates of each data point and stores them in separate lists.
    yes.append(x[1])
ax_A.scatter(xes,yes,marker="s", color='green',label='Pareto Pairs') #plots the xes list on the x-axis and yes list on the y-axis.

# limits
ax_A.plot([0,w1bar],[0,0],lw=2,color='black')
ax_A.plot([0,w1bar],[w2bar,w2bar],lw=2,color='black')
ax_A.plot([0,0],[0,w2bar],lw=2,color='black')
ax_A.plot([w1bar,w1bar],[0,w2bar],lw=2,color='black')

#adjusts the limits to make them pretty
ax_A.set_xlim([-0.1, w1bar + 0.1])
ax_A.set_ylim([-0.1, w2bar + 0.1])    
ax_B.set_xlim([w1bar + 0.1, -0.1])
ax_B.set_ylim([w2bar + 0.1, -0.1])

#Store the allocations from each question
x1A_optimal_4a, x2A_optimal_4a = model.optimal_allocation_4a() #4a
x1A_optimal_4b, x2A_optimal_4b = model.optimal_allocation_4b() #4b
x1A_optimal_5a, x2A_optimal_5a = model.optimal_allocation_5a() #5a
x1A_optimal_5b, x2A_optimal_5b = model.optimal_allocation_5b() #5b
x1A_optimal_6a, x2A_optimal_6a = model.optimal_allocation_6a() #6a


#Plot the allocations from each question:
ax_A.scatter(x1A_optimal_4a,x2A_optimal_4a,marker='s',color='red',label='4a') #plots the allocation 4a 
ax_A.scatter(x1A_optimal_4b,x2A_optimal_4b,marker='s',color='yellow',label='4b') #plots the allocation 4b
ax_A.scatter(x1A_optimal_5a,x2A_optimal_5a,marker='s',color='purple',label='5a') #plots the allocation 5a
ax_A.scatter(x1A_optimal_5b,x2A_optimal_5b,marker='s',color='turquoise',label='5b') #plots the allocation 5b
ax_A.scatter(x1A_optimal_6a,x2A_optimal_6a,marker='s',color='blue',label='6a') #plots the allocation 6a 

#Adds the describing legend to the figure:
ax_A.legend(frameon=True,loc='upper right',bbox_to_anchor=(1.6,1.0));
```



***Comparison***


When A is the market maker, A allocates the maximum amount of the goods to him self that is possible as long as B is not worse of than in the initial endowment. 


It is better for B when the consumption levels are chosen by a utilitarian social planner, as he then has a higher consumption than his initial endowment. A also has a higher consumption than initial endowment but lower than when he is the market maker.
